```python
df_orders = pd.read_csv(f'{DATA_DIR}/olist_orders_dataset.csv')
df_customers = pd.read_csv(f'{DATA_DIR}/olist_customers_dataset.csv')
df_items = pd.read_csv(f'{DATA_DIR}/olist_order_items_dataset.csv')
df_products = pd.read_csv(f'{DATA_DIR}/olist_products_dataset.csv')
df_payments = pd.read_csv(f'{DATA_DIR}/olist_order_payments_dataset.csv')
df_reviews = pd.read_csv(f'{DATA_DIR}/olist_order_reviews_dataset.csv')
df_translation = pd.read_csv(f'{DATA_DIR}/product_category_name_translation.csv')
df_sellers = pd.read_csv(f'{DATA_DIR}/olist_sellers_dataset.csv')
df_geo = pd.read_csv(f'{DATA_DIR}/olist_geolocation_dataset.csv')
```

```python
# 3.1. Bảng Orders (Xử lý Datetime & Lọc lỗi logic)
datetime_cols = ['order_purchase_timestamp', 'order_approved_at',
                 'order_delivered_carrier_date', 'order_delivered_customer_date',
                 'order_estimated_delivery_date']
for col in datetime_cols:
    df_orders[col] = pd.to_datetime(df_orders[col], errors='coerce')

df_orders['order_year'] = df_orders['order_purchase_timestamp'].dt.year
df_orders['order_month'] = df_orders['order_purchase_timestamp'].dt.month
df_orders['order_quarter'] = df_orders['order_purchase_timestamp'].dt.quarter

# Bỏ đơn hàng có ngày giao iễn ra trước ngày đặt
invalid_dates = df_orders['order_delivered_customer_date'] < df_orders['order_purchase_timestamp']
df_orders = df_orders[~invalid_dates]

print(f"Kích thước bảng Orders: {df_orders.shape}")
```

```python
# 3.2. Bảng Products
df_products = df_products.merge(df_translation, on='product_category_name', how='left')
df_products['product_category_name'] = df_products['product_category_name_english'].fillna('other')
df_products.drop(columns=['product_category_name_english'], inplace=True)
df_products['product_category_name'] = df_products['product_category_name'].str.lower().str.strip()

print(f"Kích thước bảng Products: {df_products.shape}")
```

```python
# 3.3. Xóa Duplicates theo Khóa chính
df_orders.drop_duplicates(subset=['order_id'], inplace=True)
df_customers.drop_duplicates(subset=['customer_id'], inplace=True)
df_items.drop_duplicates(subset=['order_id', 'order_item_id'], inplace=True)
df_sellers.drop_duplicates(subset=['seller_id'], inplace=True)
```

```python
# 3.4. Bảng Reviews (Giữ review mới nhất, điền null)
df_reviews = df_reviews.sort_values('review_creation_date').drop_duplicates(subset=['order_id'], keep='last')
df_reviews['review_comment_message'] = df_reviews['review_comment_message'].fillna('No Comment')
```

```python
# 3.5. Bảng Geolocation (Khử trùng lặp vĩ độ/kinh độ cho mỗi zip code)
df_geo = df_geo.drop_duplicates(subset=['geolocation_zip_code_prefix'], keep='first')
```

```python
# 4. DATA MERGING (HỢP NHẤT DỮ LIỆU)
df_merged = df_orders.merge(df_customers, on='customer_id', how='inner')
df_merged = df_merged.merge(df_items, on='order_id', how='inner')
df_merged = df_merged.merge(df_products, on='product_id', how='left')
df_merged = df_merged.merge(df_sellers, on='seller_id', how='left')

# Group bảng Payments trước khi merge để tránh nhân bản dòng
payment_grouped = df_payments.groupby('order_id').agg(
    total_payment_value=('payment_value', 'sum'),
    payment_type=('payment_type', lambda x: x.mode()[0] if not x.empty else 'unknown')
).reset_index()

df_merged = df_merged.merge(payment_grouped, on='order_id', how='left')
df_merged = df_merged.merge(df_reviews[['order_id', 'review_score']], on='order_id', how='left')

# Lọc bỏ outlier: Đơn hàng có giá trị thanh toán <= 0
df_merged = df_merged[df_merged['total_payment_value'] > 0]
print(f"Kích thước bảng Denormalized Master: {df_merged.shape}")
```

```python
df_merged.to_csv(f'{OUTPUT_DIR}/01_denormalized_master_data.csv', index=False)
df_orders.to_csv(f'{OUTPUT_DIR}/cleaned_orders.csv', index=False)
df_customers.to_csv(f'{OUTPUT_DIR}/cleaned_customers.csv', index=False)
df_items.to_csv(f'{OUTPUT_DIR}/cleaned_items.csv', index=False)
df_products.to_csv(f'{OUTPUT_DIR}/cleaned_products.csv', index=False)
df_sellers.to_csv(f'{OUTPUT_DIR}/cleaned_sellers.csv', index=False)
df_geo.to_csv(f'{OUTPUT_DIR}/cleaned_geolocation.csv', index=False)
df_payments.to_csv(f'{OUTPUT_DIR}/cleaned_payments.csv', index=False)
df_reviews.to_csv(f'{OUTPUT_DIR}/cleaned_reviews.csv', index=False)

print("Xong")
```
